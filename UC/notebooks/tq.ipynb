```python
ds[0]
```

```python
def make_all_context(sample):
    all_context = ''
    for document in sample['ctxs']:
        all_context += document['text']
        all_context += '\n\n'

    sample['all_context'] = all_context

    return sample
```

```python
ds = ds.map(make_all_context)
```

```python
ds.save_to_disk('data/datasets/trivia_qa')
```
